```python
x=np.arange(1,np.pi*3,0.1)
sin = np.sin(x)
cos=np.cos(x)
```

```python
plt.figure(figsize=(12,9))
plt.plot(sin,marker='h',linestyle = '--',alpha = 0.4)
plt.plot(cos)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin va cos funksiyalari',color='r')
plt.legend(['sin','cos'])
plt.xticks(rotation=90,size=12,color='r')
plt.grid()
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

```python
plt.figure(figsize=(16, 6))

sns.histplot(data=df, x='price')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(data=df, x='price',aspect =1 , height =10)

plt.show()
```

```python
sns.displot(data=df,x='price' ,hue='transmission',aspect = 1, height = 5,multiple = 'stack',kde=True)

plt.show()
```

```python
sns.kdeplot(data=df, x='price', shade = True,hue = 'transmission');
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```
